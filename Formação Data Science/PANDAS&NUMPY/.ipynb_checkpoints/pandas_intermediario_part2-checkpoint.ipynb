## Seleções e Frequçências

```python
import pandas as pd
dados = pd.read_csv('aluguel_residencial.csv', sep=';')
dados.head(10)
```

```python
#Selecione somente os imóveis classificados com tipo "Apartamento"

selecionados = dados['Tipo'] == 'Apartamento'
selecionados
n1 = dados[selecionados].shape[0]
n1
```

```python
#Selecione os imóveis classficados com tipo 'Casa, 'Casa de condominio e 'Casa de vila'

selecionados1 = (dados['Tipo'] == 'Casa') | (dados.Tipo =='Casa de Condominio') | (dados.Tipo == 'Casa de Vila')
selecionados1
n2 = dados[selecionados1].shape[0]
n2
```

```python
#Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites

selecionados2 = (dados['Area'] >= 40) & (dados['Area'] <= 100)
selecionados2
n3 = dados[selecionados2].shape[0]
n3
```

```python
#Selecione os imóveis que tenham pelo menos 4 quartos e aluguel
#menor do que R$ 2.000,00
selecionados3 = (dados['Quartos'] >= 4) & (dados.Valor < 2000)
selecionados3
n4 = dados[selecionados3].shape[0]
n4
```

```python
print('N° de imóveis classificados com tipo "Apartamento" é {}'.format(n1))
print('N° de  imóveis classficados com tipo "Casa", "Casa de condominio" e "Casa de vila" é {}'.format(n2))
print('N° de  imóveis com área entre 60 e 100 metros quadrados, incluindo os limites é {}'.format(n3))
print('N° de  imóveis que tenham pelo menos 4 quartos e aluguel é {}'.format(n4))
```

# Relatório de Análise 6

```python
import pandas as pd
dados = pd.read_csv('aluguel_residencial.csv', sep=';')
dados.head()
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados.IPTU
dados.head()
```

```python
dados['Valor m²'] = dados['Valor'] / dados['Area']
dados.head()
```

```python
#arredondar

dados['Valor m²'] = dados['Valor m²'].round(2)
dados
```

```python
dados['Valor Bruto m²'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head()
```

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']

dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados
```

```python
#excluir variáveis do dataframe

dados_auxiliar = pd.DataFrame(dados[['Tipo Agregado','Valor m²','Valor Bruto','Valor Bruto m²']])
dados_auxiliar
```

```python
dados.columns
```

```python
dados_auxiliar = pd.DataFrame(dados[['Tipo Agregado','Valor m²','Valor Bruto','Valor Bruto m²']])
dados_auxiliar
```

```python
del dados_auxiliar['Valor Bruto m²']
```

```python
dados_auxiliar
```

```python
dados_auxiliar.pop('Valor m²')
```

```python
dados_auxiliar
```

```python
#voltando p df original

dados.drop(['Valor Bruto', 'Valor Bruto m²'], axis = 1, inplace = True)
```

# estatísticas descritivas

```python
dados.Valor.mean()
```

```python
bairros = ['Barra da Tijuca','Copacabana','Ipanema','Leblon','Botafogo','Flamengo']
selecionados = dados['Bairro'].isin(bairros)
dados = dados[selecionados]
```

```python
dados.Bairro.drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, data.Valor.mean()))
```

```python
#maneira mais simples e mais fácil

grupo_bairro['Valor'].mean().round(2)
```

```python
grupo_bairro.Valor.describe().round(2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('figure', figsize=(10,5))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do aluguel')
fig.set_title('Valor médio do aluguel por bairro', {'fontsize': 22})
```

```python
fig = grupo_bairro['Valor'].max().plot.bar(color = 'blue')
fig.set_ylabel('Valor do aluguel')
fig.set_title('Valor médio do aluguel por bairro', {'fontsize': 22})
```
